## Flight price prediction EDA and Feature Engineering. 



```python
## Importing liberaries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## Importing training dataset. 

train_df = pd.read_excel('Data_Train.xlsx')
train_df.head()
```

```python
## Reading test dataset. 

test_df = pd.read_excel('Test_set.xlsx')
test_df.head()
```

```python
## Combining both datasets. 

final_df = pd.concat([train_df,test_df])
final_df.head()
```

```python
final_df.info()
```

### Feature Engineering - 

```python
## Breaking date of journey into 3 features. 

final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
## Converting there data typr into int

final_df['Date'] = final_df['Date'].astype(int)
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
```

```python
## Now, cheking data types. 

final_df.info()
```

```python
## Dropping Data of journey from dataset. 

final_df.drop('Date_of_Journey',axis = 1,inplace = True)
```

```python
final_df.head(1)
```

```python
## Now, Handling Feature arrival time
## splitting feature according to space. 

final_df['Arrival_Time'] = final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df.head()
```

```python
## Now splitting arrival time feature according to hour and minutes.

final_df['Arrival_hour'] = final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_min'] = final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df.head(1)
```

```python
## Now, converting there data type into integers. 

final_df['Arrival_hour'] = final_df['Arrival_hour'].astype(int)
final_df['Arrival_min'] = final_df['Arrival_min'].astype(int)
```

```python
final_df.info()
```

```python
## now dropping arrival time feature. 

final_df.drop('Arrival_Time',axis = 1,inplace = True)
```

```python
final_df.head(1)
```

```python
## Handling Feature departure time. 

final_df['Dep_hour'] = final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_min'] = final_df['Dep_Time'].str.split(':').str[1]
```

```python
final_df.head(1)
```

```python
## Changing there data type into integers, 

final_df['Dep_hour'] = final_df['Dep_hour'].astype(int)
final_df['Dep_min'] = final_df['Dep_min'].astype(int)
final_df.info()
```

```python
## Now, dropping departure time feature.

final_df.drop('Dep_Time',axis = 1,inplace = True)
```

```python
final_df.head(1)
```

```python
## handling Total_stops feature

final_df['Total_Stops'].unique()
```

```python
## Changing values of total stop feature using map function.
### This is a kind of label encoding. 
### get_dummies method is kind 

final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
final_df.head(1)
```

```python
## Dropping Route feature, since we have other features to provide necessary info. 

final_df.drop('Route',axis = 1,inplace = True)
final_df.head(1)
```

```python
## Handling Duration Feature. by splitting

final_df['Duration_hour'] = final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
final_df['Duration_min'] = final_df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
final_df.head(1)
```

```python
## There are records of 5 minutes . so, we have to drop those records. 

final_df[final_df['Duration_hour']=='5m']
```

```python
final_df.head()
```

```python
## Also , there is nan value in duration minutes, 
## So, we will convert it to 0 using fillna method

final_df['Duration_min'] = final_df['Duration_min'].fillna(0)
```

```python
final_df.head()
```

```python
final_df.drop(6474,axis = 0,inplace = True)
final_df.drop(2660,axis = 0,inplace = True)
```

```python
## Now, changing data type of duration hour to integers. 

final_df['Duration_hour'] = final_df['Duration_hour'].astype(int)
final_df['Duration_min'] = final_df['Duration_min'].astype(int)
```

```python
## now, dropping duration column

final_df.drop('Duration',axis = 1, inplace = True)
```

```python
final_df.head()
```

```python
### Handling Airline Feature. 

## We will use label encoding for this feature. 
```

```python
final_df['Airline'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
```

```python
final_df['Airline'] = labelencoder.fit_transform(final_df['Airline'])
final_df['Source'] = labelencoder.fit_transform(final_df['Source'])
final_df['Destination'] = labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info'] = labelencoder.fit_transform(final_df['Additional_Info'])
```
